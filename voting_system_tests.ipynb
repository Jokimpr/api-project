# Online Voting System API Testing Notebook

This notebook demonstrates how to test the Online Voting System endpoints programmatically using Python requests. It covers user registration, login, voting, and admin functions.

```python
import requests
from bs4 import BeautifulSoup

BASE_URL = "http://localhost:5000"

# Create a session to maintain cookies
session = requests.Session()
```

## Test User Registration

```python
# Test user registration
registration_data = {
    'email': '23bcnb31@kristujayanti.com',
    'password': 'password123',
    'confirm_password': 'password123'
}

response = session.post(f"{BASE_URL}/register", data=registration_data)
print(f"Registration Status: {response.status_code}")
print(f"Response: {response.text[:200]}...")

# Check if redirected to login
if response.url.endswith('/login'):
    print("✅ Registration successful!")
else:
    print("❌ Registration failed")
```

## Test User Login

```python
# Test user login
login_data = {
    'voter_id': '23bcnb31',
    'password': 'password123'
}

response = session.post(f"{BASE_URL}/login", data=login_data)
print(f"Login Status: {response.status_code}")
print(f"Redirect URL: {response.url}")

if '/vote' in response.url:
    print("✅ Login successful!")
else:
    print("❌ Login failed")
```

## Test Voting

```python
# Get vote page to find candidate IDs
vote_page = session.get(f"{BASE_URL}/vote")
soup = BeautifulSoup(vote_page.text, 'html.parser')

# Find candidate radio buttons
candidates = soup.find_all('input', {'name': 'candidate_id'})
if candidates:
    candidate_id = candidates[0]['value']  # Select first candidate
    print(f"Selected candidate ID: {candidate_id}")
    
    # Cast vote
    vote_data = {'candidate_id': candidate_id}
    response = session.post(f"{BASE_URL}/vote", data=vote_data)
    print(f"Vote Status: {response.status_code}")
    
    if '/results' in response.url:
        print("✅ Vote successful!")
    else:
        print("❌ Vote failed")
else:
    print("No candidates found")
```

## Test Admin Login

```python
# Test admin login (using separate session)
admin_session = requests.Session()

admin_login_data = {
    'username': 'admin',
    'password': 'admin123'
}

response = admin_session.post(f"{BASE_URL}/admin/login", data=admin_login_data)
print(f"Admin Login Status: {response.status_code}")

if '/admin/dashboard' in response.url:
    print("✅ Admin login successful!")
else:
    print("❌ Admin login failed")
```

## Check Prometheus Metrics

```python
# Check Prometheus metrics
metrics_response = requests.get(f"{BASE_URL}/metrics")
print("Prometheus Metrics:")
print(metrics_response.text[:500] + "..." if len(metrics_response.text) > 500 else metrics_response.text)
```
